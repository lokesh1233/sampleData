```python
from chatterbot import ChatBot

chatbot = ChatBot(
    'Lokesh',
    trainer='chatterbot.trainers.ChatterBotCorpusTrainer'
)

# Train based on the english corpus
chatbot.train("chatterbot.corpus.english")
```
